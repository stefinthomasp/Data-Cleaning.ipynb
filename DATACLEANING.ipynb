```python
import pandas as pd
```

```python
## reading data from csv file and storing in a variable called 'crime_boston'
crime_boston = pd.read_csv("Boston Crime.csv")
crime_boston
```

```python
## find out any missing data
crime_boston.isnull()

## 'False' shows no missing while 'True' shows missing
```

```python
## 'retreives the rows' having the  Missing data ; we use the method 'any()'

crime_boston.isnull().any(axis=1)
```

```python
## retreive the rows from the data having 'missing values'

rows_with_missing_values = crime_boston.isnull().any(axis=1)
crime_boston[rows_with_missing_values]
```

```python
## In the above output ,the missing values in  a row contains the value ' NaN'
```

```python
## drop- cleaning the dataset may lead to the accurate results.
## Drop the columns 'Year', ' Month' and 'Hour' because it is already contained in the column named 'Occured on Date'

cleaned_data = crime_boston.drop(columns=['YEAR','MONTH','HOUR'])
```
